```python
import sys,os
sys.path.append(os.path.dirname(os.path.abspath('')))
from ioMicro import *
```

```python
save_folder =r'\\192.168.0.6\bbfish1e3\DCBBL1_03_14_2023_big\MERFISH_Analysis'
fld_aso = r'\\192.168.0.3\bbfishdc9\DCBBL1_03_14_2023_big\P3_Ptbp2_ASO_PolyA'

fov,set_ = 'Conv_zscan__582', '_set2'
dec = decoder_simple(save_folder,fov,set_)
save_fl_aso = dec.save_folder+os.sep+'mean_aso_'+dec.fov+'--'+dec.set_+'.npz'
if not os.path.exists(save_fl_aso):
    load_segmentation(dec)
    fl_aso = fld_aso+dec.set_+os.sep+os.path.basename(dec.fl_ref)
    im_ASO = read_im(fl_aso)
    resc = (dec.shape/dec.im_segm_.shape).astype(int)
    im_ASOf = im_ASO[1][::resc[0],::resc[1],::resc[2]]
    im_ASO_dapi = im_ASO[-1][::resc[0],::resc[1],::resc[2]]
    im_dapi_ref = read_im(dec.fl_ref)[-1][::resc[0],::resc[1],::resc[2]]
    im_polyt = im_ASO[2][::resc[0],::resc[1],::resc[2]]
    im_ASO_dapi = np.array(im_ASO_dapi,dtype=np.float32)
    im_dapi_ref = np.array(im_dapi_ref,dtype=np.float32)
    txyz,txyzs = get_txyz(im_ASO_dapi,im_dapi_ref,sz=300,sz_norm=30)
    im_segm_ = nd.shift(dec.im_segm_,-txyz,order=0)
    if False:
        v = napari.view_labels(im_segm_)
        v.add_image(im_ASO_dapi)#,translate=txyz)
    cells = np.unique(dec.im_segm_)
    mean_aso = nd.mean(np.array(im_ASOf,dtype=np.float32),im_segm_,cells)
    mean_polyt = nd.mean(np.array(im_polyt,dtype=np.float32),im_segm_,cells)
    np.savez(save_fl_aso,mean_aso=mean_aso,mean_polyt=mean_polyt,cells=cells)
```
