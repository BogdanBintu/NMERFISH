```python
def get_meds(tag = '_set4',folder = r'\\192.168.0.96\bbfish124\HttHuman_Snonia__06_12_2023'):
    save_fls = glob.glob(folder+r'\CAG_quant\*'+tag+'*.npz')
    meds = []
    meds_r = []


    for fl in save_fls:
        dic = np.load(fl)
        h_CAG_ = dic['h_CAG_']
        h_HTT_ = dic['h_HTT_']
        Xh = dic['XhHTT']
        keep = Xh[:,0]>3 #off the coverslip
        keep = keep&(Xh[:,-2]>0.7)  #high quality Htt

        meds.append(np.mean(np.log(h_CAG_[keep])))
        meds_r.append(np.mean(np.log(h_CAG_[keep])-np.log(h_HTT_[keep])))
    return np.exp(meds),np.exp(meds_r)
#np.median(meds),np.median(meds_r)
#plt.plot(np.exp(meds))
```

```python
def check_image(index=0,tag = '_set4',folder = r'\\192.168.0.96\bbfish124\HttHuman_Snonia__06_12_2023',
               raw_data_folder = r'\\192.168.0.96\bbfish124\HttHuman_Snonia__06_12_2023\P3__Htt_CAG_B_polyA_'):
    save_fls = glob.glob(folder+r'\CAG_quant\*'+tag+'*.npz')
    save_fl  = save_fls[index]
    #save_fl = save_fls[0]
    fov = os.path.basename(save_fl).split('--')[0]
    set_ = os.path.basename(save_fl).split('--')[-1].split('__CAG')[0]
    fl = raw_data_folder+set_+os.sep+fov+'.zarr'
    im = read_im(fl)
    import napari
    napari.view_image(im)
    viewer.add_points(X,face_color=[0,0,0,0],edge_color='w',size=20)
```
